```python
#Q.2 on ={'+','-','/','*'}
class UserMainCode(object):
    @classmethod
    def letter(cls,input1):
        '''
        input1:string
        Expected return type: int
        '''
        s=input1
        n=len(s)
        stack=[]
        a={'+','-','/','*'}
        for i in range(n):
            if s[i].isdigit():
                stack.append(int(s[i]))
            elif s[i]=='+':
                op1=stack.pop()
                op2=stack.pop()
                stack.append(int(op2)+int(op1))
            elif s[i]=='-':
                op1=stack.pop()
                op2=stack.pop()
                stack.append(int(op2)-int(op1))
            elif s[i]=='/':
                op1=stack.pop()
                op2=stack.pop()
                stack.append(int(op2)/int(op1))
            elif s[i]=='*':
                op1=stack.pop()
                op2=stack.pop()
                stack.append(int(op2)*int(op1))
            return stack.pop()
```

```python
#Q.4 sum of letters
class UserMaincode(object):
    @claassmethod
    def letter(cls, input1):
        '''
        input1 : string
    
        Expected return type : int
        '''
        
        # Read only region end
        # Write code here
        scores = {}
        scores['A'] = 0
        scores['B'] = 1
        for j in range(ord('A')+2, ord(('Z')+1)):
            scores[chr(j)] = scores[chr(j-1) + scores[chr(j-2)]]
        total = 0
        for j in input1:
            total += scores[j]
        return total
```
